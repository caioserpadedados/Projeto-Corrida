```python
import matplotlib.pyplot as plt
import seaborn as sns

# Dados
categorias = ["Clubes de Corrida (Brasil)", "Média Global", "Maratonas e Ultramaratonas"]
percentuais = [109, 59, 9]

# Criando um DataFrame para associar as cores
import pandas as pd
df = pd.DataFrame({"Categoria": categorias, "Crescimento": percentuais})

# Configuração do gráfico
plt.figure(figsize=(8, 5))
sns.barplot(x="Categoria", y="Crescimento", data=df, palette=["blue", "gray", "green"], hue="Categoria")

# Adicionando rótulos
plt.ylabel("Crescimento (%)")
plt.title("Crescimento da Corrida de Rua em 2024")
plt.ylim(0, 120)

# Exibir valores no topo das barras
for i, v in enumerate(percentuais):
    plt.text(i, v + 2, f"{v}%", ha='center', fontsize=12, fontweight='bold')

plt.show()
```
